加载鸢尾花数据集并展示基本信息  

对特征进行标准化处理  

划分训练集和测试集

训练高斯朴素贝叶斯模型（基于贝叶斯定理的分类算法）

模型预测与评估（准确率、混淆矩阵、分类报告）  

展示预测概率（贝叶斯模型的核心输出）  

可视化决策边界（使用前两个特征）  

展示每个特征在不同类别中的概率分布（直观理解贝叶斯估计的原理）  

使用训练好的模型对新数据进行预测并输出预测概率  

高斯朴素贝叶斯的核心思想是：假设特征之间相互独立，通过计算后验概率（P (类别 | 特征)）来进行分类，选择后验概率最大的类别作为预测结果。

```python
# 导入必要的库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap

# 设置中文显示
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

# 1. 加载数据集
iris = load_iris()
X = iris.data  # 特征数据
y = iris.target  # 标签数据

# 查看数据集信息
print("特征名称:", iris.feature_names)
print("标签名称:", iris.target_names)
print("数据集形状:", X.shape)

# 2. 数据预处理 - 特征标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42  # 30%作为测试集
)

print(f"训练集大小: {X_train.shape[0]}")
print(f"测试集大小: {X_test.shape[0]}")

# 4. 训练高斯朴素贝叶斯模型
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# 5. 模型预测
y_pred = gnb.predict(X_test)
y_pred_proba = gnb.predict_proba(X_test)  # 预测概率

# 6. 模型评估
print("\n模型准确率:", accuracy_score(y_test, y_pred))
print("\n混淆矩阵:\n", confusion_matrix(y_test, y_pred))
print("\n分类报告:\n", classification_report(
    y_test, y_pred, target_names=iris.target_names
))

# 查看前5个样本的预测概率
print("\n前5个测试样本的预测概率:")
for i in range(5):
    print(f"样本{i+1}: {y_pred_proba[i].round(4)}, 真实标签: {iris.target_names[y_test[i]]}, 预测标签: {iris.target_names[y_pred[i]]}")

# 7. 可视化决策边界（使用前两个特征）
# 为了可视化，使用前两个特征重新训练模型
X_vis = X_scaled[:, :2]  # 只取前两个特征
X_train_vis, X_test_vis, y_train_vis, y_test_vis = train_test_split(
    X_vis, y, test_size=0.3, random_state=42
)

# 训练可视化用的模型
gnb_vis = GaussianNB()
gnb_vis.fit(X_train_vis, y_train_vis)

# 创建网格以绘制决策边界
h = 0.02  # 网格步长
x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# 预测网格点的类别
Z = gnb_vis.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 绘制决策边界和数据点
plt.figure(figsize=(10, 6))
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.8)
plt.scatter(X_train_vis[:, 0], X_train_vis[:, 1], c=y_train_vis, cmap=cmap_bold, edgecolor='k', label='训练集')
plt.scatter(X_test_vis[:, 0], X_test_vis[:, 1], c=y_test_vis, cmap=cmap_bold, edgecolor='red', marker='*', s=100, label='测试集')

plt.title('高斯朴素贝叶斯决策边界')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.show()

# 8. 贝叶斯模型的概率分布可视化（每个类别的特征分布）
plt.figure(figsize=(12, 8))
for i in range(4):  # 遍历4个特征
    plt.subplot(2, 2, i+1)
    for target, color in zip(range(3), ['r', 'g', 'b']):
        # 绘制每个类别的特征分布直方图
        plt.hist(X_scaled[y == target, i], bins=15, alpha=0.5, label=iris.target_names[target], color=color)
    plt.xlabel(iris.feature_names[i])
    plt.ylabel('频率')
    plt.legend()

plt.tight_layout()
plt.suptitle('各特征在不同类别中的分布', y=1.02)
plt.show()

# 9. 使用模型进行新数据预测
new_sample = np.array([[5.1, 3.5, 1.4, 0.2]])  # 类似山鸢尾的特征
new_sample_scaled = scaler.transform(new_sample)  # 标准化新样本
prediction = gnb.predict(new_sample_scaled)
prediction_proba = gnb.predict_proba(new_sample_scaled)

print(f"\n新样本预测结果: {iris.target_names[prediction][0]}")
print(f"预测概率: {prediction_proba.round(4)}")
```
